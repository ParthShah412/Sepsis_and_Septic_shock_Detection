```python
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import shutil

import wfdb
```

```python
# Demo 1 - Read a wfdb record using the 'rdrecord' function into a wfdb.Record object.
# Plot the signals, and show the data.
record = wfdb.rdrecord("3544749_0001") 
wfdb.plot_wfdb(record=record, title='3544749_0001') 
#display(record.__dict__)

record2 = wfdb.rdrecord("3544749_0002") 
wfdb.plot_wfdb(record=record2, title='3544749_0002') 
#display(record.__dict__)

record3 = wfdb.rdrecord("3544749_0003") 
wfdb.plot_wfdb(record=record3, title='3544749_0003') 

record4 = wfdb.rdrecord("3544749_0004") 
wfdb.plot_wfdb(record=record4, title='3544749_0004') 

recordlongn = wfdb.rdrecord("p000020-2183-04-28-17-47n")
wfdb.plot_wfdb(record= recordlongn, title= "longn")

recordlong = wfdb.rdrecord("p000020-2183-04-28-17-47")
wfdb.plot_wfdb(record= recordlong, title= "long")
```
